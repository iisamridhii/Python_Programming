3.Deep and Shallow copy

shallow copy --no changes occur in main array even after performing opertions on it

```python
a= np.arange(1,21)
a
```

```python
slice=a[:5]
slice
```

```python
slice=slice*10
slice
```

```python
a #no changes in a
```

deep copy-- changes do occur in the main array too

```python
a
b=a
```

```python
b[0]=99
b
```

```python
a #Changes happened in the main array too
```

```python
#changes made in b also occured in a this concept is called deep copy
```

```python
a
b=a.copy() #SHALLOW COPY
```

```python
b[0]=99
b
```

```python
a
```
