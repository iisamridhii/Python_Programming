NUMPY INDEXING AND SLICING

```python
arr=np.arange(11,21)
arr
```

```python
arr[6] #GIVES THE ELEMENT PRESENT AT THAT GIVEN INDEX
```

```python
arr[1]
```

```python
arr[0]
```

```python
arr[1:4] #Slicing 
```
