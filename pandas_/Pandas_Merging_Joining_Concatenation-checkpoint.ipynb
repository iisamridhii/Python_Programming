# Merging 2 Dataframes

```python
employees=pd.DataFrame({           #DataFrame1 employee info.
    'employee_id': [1,2,3,4,5],
    'name': ['Jhon','Anna','Peter','Linda','Bob'],
    'department': ['HR','IT','Finance','IT','HR']
})
#DataFrame2 Salary info.
salaries = pd.DataFrame({
    'employee_id': [1,2,3,6,7],
    'salary': [60000,80000,65000,70000,90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries)
```

```python
pd.merge(employees, salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees, salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees, salaries,on='employee_id',how='left') #here employees is in left & salaries is in right
#So left gives preference to employes over salaries here
```

```python
pd.merge(employees, salaries,on='employee_id',how='right') #right-- preference to salaries over employees
```

# Concatenation of 2 DataFrames

### Joining of 2 Dataframes

```python
df1= pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2'],
    'C': ['C0','C1','C2']
})

df2= pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B': ['B3','B4','B2'],
    'C': ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #takes a list only
#column
```

```python
pd.concat([df1,df2],axis=1) #row
```

# Joining 2 dataframes

```python
df1= pd.DataFrame({
    'name': ['Alice','Bob','Charlie'],
}, index =[1,2,3])

#2nd DataFrame
df2 = pd.DataFrame({
    'score': [85,90,75]
}, index=[2,3,4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how='outer')
```
