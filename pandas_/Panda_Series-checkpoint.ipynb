## PANDA SERIES

A series is a one dimensional labeled array capable of holding any data type. The axis labels are collectively called the index.

# Creating Series

There are multiple ways to create a pandas Series:

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c'] #labels
my_list = [10,20,30] #list
arr=np.array([10,20,30]) #array 1-D
d= {1:10,2:20,3:30} #dictionary
          
```

```python
pd.Series(my_list)
```

here, 0 1 2 are Labels and 10 20 30 are data points

```python
pd.Series(my_list,index=labels)
```

here we had set the labels to a b c via labels variable above

series cannot take 2-D arrays only 1-D allowed because series itself is one dimensional!!

```python
pd.Series(arr)
```

```python
pd.Series(d) #in dictionary the keys will become labels and values will become data points
```
